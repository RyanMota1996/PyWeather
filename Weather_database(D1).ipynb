```python
# import dependances
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from citipy import citipy 
from config import weather_api_key
import requests

# Import the time library and the datetime module from the datetime library 
import time
from datetime import datetime

# setup longitudes and latitudes ranges
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
#generating random geolocations
lat_lngs = []
cities = []

lats = np.random.uniform(low=-90.000, high=90.000, size=2000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=2000)
lat_lngs = zip(lats, lngs)

#finding nearest cities

for lat_lngs in lat_lngs:
    city = citipy.nearest_city(lat_lngs[0], lat_lngs[1]).city_name
    
    #find unqiue cities
    if city not in cities:
        cities.append(city)
len(cities)
        
```

```python
city_dict={"City":cities}
city_dict_df = pd.DataFrame(city_dict)
output_data_file = "cities.csv"
city_dict_df.to_csv(output_data_file, index_label="City_ID")
```

```python
#preforming api call
new_cities = []
cloudiness= []
country = []
humidity=[]
temp=[]
lat=[]
lng=[]
wind=[]
descrip=[]
```

```python
city_data=[]
record_counter = 0
set_counter = 0
# Starting URL for Weather Map API Call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key 
print('------------------------')
print('Beginning Data Retrieval')
print('------------------------')

for city in cities:
    query_url = url + "&q=" + city
    # Get weather data
    response = requests.get(query_url).json()
    if record_counter < 50:
        record_counter += 1
    else:
        set_counter += 1
        record_counter = 0

    print(f"Processing Record {record_counter} of Set {set_counter} | {city}")
    print(query_url)
    try:
        cloudiness.append(response['clouds']['all'])
        country.append(response['sys']['country'])
        humidity.append(response['main']['humidity'])
        temp.append(response['main']['temp_max'])
        lat.append(response['coord']['lat'])
        lng.append(response['coord']['lon'])
        wind.append(response['wind']['speed'])
        descrip.append(response["weather"][0]["description"])
        new_cities.append(city)
        
    except:
        print("City not found. Skipping...")
        pass

print('-------------------------')
print('Data Retrieval Complete')
print('-------------------------')
```

```python

weather_dict = {
    "Country" : country,
    "City": new_cities,
    "Cloudiness (%)" : cloudiness,
    "Humidity (%)" : humidity,
    "Max Temperature": temp,
    "Latitude (Deg)" : lat,
    "Longitude (Deg)" : lng,   
    "Wind Speed (mph)" : wind,
    "Description" : descrip
}
weather_df = pd.DataFrame.from_dict(weather_dict, orient='index')
weather_df=weather_df.transpose()
```
